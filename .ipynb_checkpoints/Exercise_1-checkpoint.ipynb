```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal, norm
from mpl_toolkits.mplot3d import Axes3D # <--- This is important for 3d plotting 

np.random.seed(42)
```

## Exercise 1

- Implement a function that draws `1000` samples from an univariant gaussian distribution with mean `1` and a standard deviation of `0.2`. 
- Plot the experimental distribution using `30` bins. 
- In order to compare it with the ground truth distribution, generate also the plot of such.

<br>

Probability Density Function for Gaussians (PDF)

## $\displaystyle f(t) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(t-\mu)^2}{2\sigma^2}}$

<a name='1.1'></a>

### Exercise 1.1
### `Gaussian Distribution`

```python
mu = 1
sigma = 0.2
sample_size = 1000

bins = 30

samples = np.random.normal(mu, sigma, sample_size)

_, bins,_ = plt.hist(samples, bins=bins)
plt.title("Gaussian Distribution with mean 1 and std of 0.2")
plt.show()
```

<a name='1.2'></a>

### Exercise 1.2
### `PDF`

```python
pdf = 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) )

plt.plot(bins, pdf, color="r")
plt.title("Probability Density Function with mean 1 and std of 0.2")
plt.show()
```

## Exercise 2

The probability density function for multivariate normal distribution is:

$\displaystyle f(x) = \frac{1}{\sqrt{(2 \pi)^k \det \Sigma}} \exp\left( -\frac{1}{2} (x - \mu)^T \Sigma^{-1} (x - \mu) \right)$

<br>

- Implement a function that draws 10000 samples from a multivariant gaussian dis- tribution (2D in this case) with mean mean vector `[0.5, −0.2]` and covariance matrix `[[2.0, 0.3], [0.3, 0.5]]`. 
- Plot the experimental distribution using `30` bins `(30x30)`. 
- In order to compare it with the ground truth distribution, generate also the plot of such.

```python
# Initialize Parameters
mu = [0.5, -0.2]
cov = [[2.0, 0.3], [0.3, 0.5]]
bins = 30
n = 10000

# Positive Definite
np.linalg.eigvals(cov)

# Distribution
rv = multivariate_normal(mu, cov)
```

<a name='2.1'></a>

### Exercise 2.1
### `Multivariate Gaussian Distribution`

```python
def multivariate_gauss_dist(bins=30, sample_size=10000, figure=None):
    """
        INPUT: 
            - bins (int): number of bins, default = 30
            - sample_size (int): number of samples, default = 10000
            - figure (obj): For exercise 3, default False
            
        OUTPUT:
            - plot of the hist
    """
    
    sampled_data = rv.rvs(sample_size)

    x = sampled_data[:,0]  
    y = sampled_data[:,1]
    
    hist, xedges, yedges = np.histogram2d(x, y, bins=(bins,bins))
    
    x_bins = np.linspace(-6,6, bins)
    y_bins = np.linspace(-3,3, bins)
    
    xpos, ypos = np.meshgrid(x_bins, y_bins)

    xpos = xpos.flatten()
    ypos = ypos.flatten()
    
    zpos = np.zeros_like(xpos)

    dx = xedges [1] - xedges [0]
    dy = yedges [1] - yedges [0]

    dz = hist.flatten()

    if(figure == None):
        fig = plt.figure(figsize=(7,7))
        ax = fig.add_subplot(111, projection='3d')
        
        ax.bar3d(xpos, ypos, zpos, dx, dy, dz, color='#44a1db', zsort='average')
        ax.set_ylabel('Y axis')
        ax.set_xlabel('X axis')
        ax.set_zlabel('Z axis')
        ax.view_init(35, 300)
        
    if(figure):
        axs = figure.add_subplot(121, projection='3d')
        axs.bar3d(xpos, ypos, zpos, dx, dy, dz, color='#44a1db', zsort='average')
        axs.set_ylabel('Y axis')
        axs.set_xlabel('X axis')
        axs.set_zlabel('Z axis')
        axs.view_init(35, 300)
```

```python
# default case
multivariate_gauss_dist(bins=bins, sample_size=n)
```

<a name='2.2'></a>

### Exercise 2.2
### `PDF`

```python
def multivariate_gauss_pdf(bins=30, axs=None, figure=None):
    """
        INPUT:
            - bins (int): number of bins, default = 30
            - sub_ax: For exercise 3, default False

        OUTPUT:
            - Multivariate Gaussian PDF Function
    """
    x = np.linspace(-6,6, bins)
    y = np.linspace(-3,3, bins)

    X, Y = np.meshgrid(x,y)

    pos = np.empty(X.shape + (2,))

    pos[:, :, 0] = X
    pos[:, :, 1] = Y
    
    if(figure == None):
        fig = plt.figure(figsize=(7,7))
        ax = fig.gca(projection='3d')
        ax.plot_surface(X, Y, rv.pdf(pos), color='red')
        ax.set_xlabel('X axis')
        ax.set_ylabel('Y axis')
        ax.set_zlabel('Z axis')
        ax.view_init(35, 300)
        
    if(figure):
        axs = figure.add_subplot(122, projection='3d')
        axs.plot_surface(X, Y, rv.pdf(pos), color='red')
        axs.set_xlabel('X axis')
        axs.set_ylabel('Y axis')
        axs.set_zlabel('Z axis')
        axs.view_init(35, 300)
```

```python
#default case
multivariate_gauss_pdf(bins=bins)
```
